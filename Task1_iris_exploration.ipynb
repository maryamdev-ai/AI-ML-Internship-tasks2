```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
# Ensures plots appear in the notebook (comment on separate line)
```

```python
iris = sns.load_dataset('iris')  # Load built-in dataset from seaborn
iris.head()  # Show first 5 rows
```

```python
print("Shape:", iris.shape)  # (150 rows, 5 columns)
print("\nColumns:", iris.columns.tolist())
print("\nMissing values?:", iris.isnull().sum())  # Check for missing data
```

```python
iris.describe()  # Statistical summary (mean, min, max, etc.)
```

```python
iris['species'].value_counts()  # Counts per species
```

```python
sns.pairplot(iris, hue='species')
plt.suptitle("Pairplot of Iris Features", y=1.02)
plt.show()
```

```python
iris['species'].value_counts()  # Counts per species
```

```python
sns.pairplot(iris, hue='species')
plt.suptitle("Pairplot of Iris Features", y=1.02)
plt.show()
```

```python
iris.hist(figsize=(10, 6), bins=20)
plt.suptitle("Feature Distributions", y=1.02)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=iris.melt(id_vars='species'), x='variable', y='value', hue='species')
plt.title("Box Plots by Species")
plt.xticks(rotation=45)
plt.show()
```
